Python自动发送邮件主要是两个库：email、smtplib，email负责构造邮件，smtplib负责发送邮件。

- 无附件发送

```python
from email.header import Header
from email.mime.text import MIMEText
from email.utils import formataddr
import smtplib

from_addr = '3310612297@qq.com'     # 发件人邮箱账号
password_sq = 'dsircqixhiacciii'    #qq邮箱授权码,不是登录密码

to_addr = '1013356193@qq.com,'      #收件人邮箱账号，注意末尾“,”，因为sendmail收件人参数是列表形式，使用了split(',')
                                    #如若群发直接填入多个地址，用“,”隔开，例如'2483278443@qq.com,1013356193@qq.com,'

def sendMail(email_title, email_content):
    
    '''
    功能：发送邮件。
    参数：标题、内容。
    '''
    
    
    msg = MIMEText(email_content, "plain", 'utf-8')        # 发送邮件内容
    msg["Subject"] = Header(email_title, 'utf-8')           # 发送邮件主题/标题
    msg["From"] = formataddr(['皇上', from_addr])            # 邮件发送方
    #msg["To"] = formataddr(['爱妃', to_addr])          # 邮件接收方
    msg["To"] = to_addr
    print(msg)
    try:
        s = smtplib.SMTP("smtp.qq.com", 25)        # 邮箱的传输协议，端口默认25
        s.login(from_addr, password_sq)     # 登录邮箱
        s.set_debuglevel(1)   #打印出和SMTP服务器交互的所有信息
        #s.sendmail(from_addr, [to_addr, ], msg.as_string())  # 发送方，接收方，发送消息
        s.sendmail(from_addr, to_addr.split(','), msg.as_string())  # 发送方，接收方，发送消息
        s.quit()               # 退出邮箱
        print("邮件发送成功！")
    except Exception:
        print("邮件发送失败~~")
 
if __name__ == '__main__':
    
    mail_title = '皇上，该翻牌子了'
    mail_content = '爱妃，侍寝，清代的皇帝，都有勤政的传统。所以，想要睡懒觉肯定是不行的。他们大概早晨5点就要起床，起床第一件事是穿衣服。这听起来是句废话，其实不然。皇帝的衣服，自然要比凡人的复杂许多，大家看《延禧攻略》中的道具，一件皇帝的龙袍，以现在的工艺复制都价值不菲，当时更是价值连城，而穿衣服也是个细活，需要多人伺候，花费很长时间。'
    sendMail(mail_title, mail_content)
```

- 有附件发送

```python
from email.mime.multipart import MIMEMultipart 
from email.mime.base import MIMEBase
from email import encoders
from email.header import Header
from email.mime.text import MIMEText
from email.utils import formataddr
import smtplib

from_addr = '3310612297@qq.com'     # 发件人邮箱账号
password_sq = 'dsircqixhiacciii'    #qq邮箱授权码,不是登录密码

to_addr = '1013356193@qq.com,'      #收件人邮箱账号，注意末尾“,”，因为sendmail收件人参数是列表形式，使用了split(',')
                                    #如若群发直接填入多个地址，用“,”隔开，例如'2483278443@qq.com,1013356193@qq.com,'

def sendMail(email_title, email_content):
    
    '''
    功能：发送邮件。
    参数：标题、内容。
    '''
    
 
    #邮件对象
    msg = MIMEMultipart()

    msg["Subject"] = Header(email_title, 'utf-8')           # 发送邮件主题/标题
    msg["From"] = formataddr(['皇上', from_addr])            # 邮件发送方
    #msg["To"] = formataddr(['爱妃', to_addr])          # 邮件接收方
    msg["To"] = to_addr

    # 邮件正文是MIMEText:
    msg.attach(MIMEText(email_content, "plain", 'utf-8'))

    # 添加附件就是加上一个MIMEBase，从本地读取一个图片:
    with open('/Users/shining/Desktop/642f2934c38b408790e5c0d13436ac06.jpg', 'rb') as f:
        # 设置附件的MIME和文件名，这里是jpg类型:
        mime = MIMEBase('image', 'jpg', filename='test.jpg')
        # 加上必要的头信息，filename为附件名，可以随意命名
        mime.add_header('Content-Disposition', 'attachment', filename='test.jpg')
        mime.add_header('Content-ID', '<0>')
        mime.add_header('X-Attachment-Id', '0')
        # 把附件的内容读进来:
        mime.set_payload(f.read())
        # 用Base64编码:
        encoders.encode_base64(mime)
        # 添加到MIMEMultipart:
        msg.attach(mime)
        #print(msg)
        
    with open('/Users/shining/Desktop/test.txt', 'rb') as f:
        # 设置附件的MIME和文件名，这里是text类型:
        mime = MIMEBase('text', 'txt', filename='test.txt')
        # 加上必要的头信息，filename为附件名，可以随意命名
        mime.add_header('Content-Disposition', 'attachment', filename='test.txt')
        mime.add_header('Content-ID', '<0>')
        mime.add_header('X-Attachment-Id', '0')
        # 把附件的内容读进来:
        mime.set_payload(f.read())
        # 用Base64编码:
        encoders.encode_base64(mime)
        # 添加到MIMEMultipart:
        msg.attach(mime)
        #print(msg)



    try:
        s = smtplib.SMTP("smtp.qq.com", 25)        # 邮箱的传输协议，端口默认25
        s.login(from_addr, password_sq)     # 登录邮箱
        s.set_debuglevel(1)   #打印出和SMTP服务器交互的所有信息
        #s.sendmail(from_addr, [to_addr, ], msg.as_string())  # 发送方，接收方，发送消息
        s.sendmail(from_addr, to_addr.split(','), msg.as_string())  # 发送方，接收方，发送消息
        s.quit()               # 退出邮箱
        print("邮件发送成功！")
    except Exception:
        print("邮件发送失败~~")
 
if __name__ == '__main__':
    
    mail_title = '皇上，该翻牌子了'
    mail_content = '爱妃，侍寝，清代的皇帝，都有勤政的传统。所以，想要睡懒觉肯定是不行的。他们大概早晨5点就要起床，起床第一件事是穿衣服。这听起来是句废话，其实不然。皇帝的衣服，自然要比凡人的复杂许多，大家看《延禧攻略》中的道具，一件皇帝的龙袍，以现在的工艺复制都价值不菲，当时更是价值连城，而穿衣服也是个细活，需要多人伺候，花费很长时间。'
    sendMail(mail_title, mail_content)
```
