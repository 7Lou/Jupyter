```python
import requests,json
import pandas as pd
```

1.获取每个国家的名称

```python
def foreignList():
    url = 'https://view.inews.qq.com/g2/getOnsInfo?name=disease_foreign'
    response = requests.get(url)
    foreignList = json.loads(json.loads(response.text)['data'])['foreignList']
        
    return foreignList

def get_data(foreignList):
    all_data = {}
    for i in foreignList:
        url = f"https://api.inews.qq.com/newsqa/v1/query/pubished/daily/list?country={i['name']}"
        print(url)
        response = requests.post(url)
        data = json.loads(response.text)['data']
        all_data[i['name']] = data
        
    return all_data
    #print(all_data)

foreignList = foreignList()
data = get_data(foreignList)
df =pd.DataFrame(data)
print(df)
#df.to_excel('/Users/shining/Desktop/feiyan.xlsx')

    
```
