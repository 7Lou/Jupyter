```python
import requests,re
from pyquery import PyQuery as pq


def page_list(url):
    '''
    功能：获取所有章回链接。
    参数：小说网址。
    '''
    response = requests.get(url)
    #print(response.text)
    doc = pq(response.text)
    #print(doc('.c_con_list'))
    for each in doc('.c_con_li_detail').items():
        yield[each.text(),'https:'+each.attr('href')] #章回标题，章回链接
        
def save_page(chapter,href):
    '''
    功能：保存数据
    参数：章回标题、章回链接
    '''
    response = requests.get(href)
    doc = pq(response.text)
    text = doc('.noveContent').text()
    
    with open(f'/Users/shining/Documents/未命名文件夹/幻想苍神传/{chapter}.txt','w') as f:
        f.write(text)
         
if __name__ == "__main__":
    url = 'https://b.faloo.com/html/358/358952/'
    for i in page_list(url):
        print(i[0],i[1])
        chapter = i[0]
        href = i[1]
        save_page(chapter,href)
    print('ok')

```
