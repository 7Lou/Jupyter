```python
import requests,openpyxl,time
import pandas as pd

def get_page(num,ids):
    
    
    url = f'https://api.eol.cn/gkcx/api/?access_token=&admissions=&central=&department=&dual_class=&f211=&f985=&is_dual_class=&keyword=&page={num}&province_id={ids}&request_type=1&school_type=&signsafe=&size=20&sort=view_total&type=&uri=apigkcx/api/school/hotlists'
    headers={
        'user-agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_3) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.132 Safari/537.36'
    }
    try:
        response = requests.get(url,headers=headers)
        if response.status_code == 200:
            return response.json()
    except:
        return None
    
def get_school_id(json1):
    
    
    if json1.get('data'):
        school_id = []
        for item in json1.get('data').get('item'):
            school_id.append(item.get('school_id'))
            
        return school_id
    
def get_school_data(school_id):
    
    url = f'https://static-data.eol.cn/www/2.0/school/{school_id}/info.json'
    headers={
        'user-agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_3) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.132 Safari/537.36'
    }
    try:
        response = requests.get(url,headers=headers)
        
        if response.status_code == 200:
            data = response.json().get('data')
            return{
                'name': data.get('name'),
                'belong': data.get('belong'),
                'city_name': data.get('city_name'),
                'level_name':data.get('level_name'),
                'type_name': data.get('type_name'),
                'school_nature_name': data.get('school_nature_name'),
                'dual_class_name': data.get('dual_class_name'),
                'site': data.get('site'),
                'phone': data.get('phone'),
                'email': data.get('email'),
                'town_name': data.get('town_name'),
                'address': data.get('address')
            }
    except:
        return None
    
def save_data(content):
    cont_list = content
    df =pd.DataFrame(cont_list,columns=['name','belong','city_name','level_name','type_name','school_nature_name',
        'dual_class_name','site','phone','email','town_name','address'])
    #print(df)
    df.to_excel('/Users/shining/Desktop/jilin_gaoxiao.xlsx')
    print('ok')
    
    
def main():
    ids = 22
    data_list=[]
    for num in [1,2,3,4,5,6,7,8,9]:
        print(f'正在爬取{num}页')
        json1 = get_page(num,ids)
        for i in get_school_id(json1):
            school_id = i
            school_data = get_school_data(school_id)
            data_list.append(school_data)
            #print(school_data)
        time.sleep(1)
        
    #save_data(data_list)
    print(data_list)
        
            
if __name__ == '__main__':
    main()
    
    
```
